```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt

# pip3 install tensorflow
# pip3 install pandas
# pip3 install matplotlib
```

```python
df = pd.read_csv('./iris.csv')
df
```

```python
labels, uniques = pd.factorize(df['Species'])
labels, uniques
```

```python
session = tf.Session()
```

```python
features = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']
x = df[features].as_matrix().astype('float32')
x
```

```python
x -= x.min()
x /= x.max()
x
```

```python
k1 = 6
k2 = 5

w1 = tf.Variable(tf.truncated_normal([len(features), k1]))
b1 = tf.Variable(tf.truncated_normal([k1]))
y1 = tf.nn.relu(tf.matmul(x, w1) + b1)

w2 = tf.Variable(tf.truncated_normal([k1, k2]))
b2 = tf.Variable(tf.truncated_normal([k2]))
y2 = tf.nn.relu(tf.matmul(y1, w2) + b2)

w = tf.Variable(tf.truncated_normal([k2, len(uniques)]))
b = tf.Variable(tf.truncated_normal([len(uniques)]))
y = tf.nn.softmax(tf.matmul(y2, w) + b)

z = tf.one_hot(labels, 3)

cross_entropy = -tf.reduce_sum(z * tf.log(y), axis=1)
loss = tf.reduce_mean(cross_entropy)

train = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
```

```python
correct = tf.equal(tf.argmax(y, axis=1), labels)
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```

```python
session.run(tf.global_variables_initializer())

def draw(index):
    print(index, session.run(accuracy))
    c = session.run(tf.argmax(y, axis=1))
    plt.scatter(x[:, 0], x[:, 1], c=['rgb'[color] for color in c])
    plt.show()

draw(0)

for i in range(100):
    for j in range(5):
        session.run(train)
    draw(i + 1)
```
