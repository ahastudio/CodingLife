```python
labels, uniques = pd.factorize(df['Species'])
labels, uniques
```

```python
x = df['Sepal.Length'].as_matrix()
y = df['Sepal.Width'].as_matrix()

plt.scatter(x, y)
plt.show()
```

```python
for i in range(len(uniques)):
    data = df[df['Species'] == uniques[i]]
    x = data['Sepal.Length'].as_matrix()
    y = data['Sepal.Width'].as_matrix()
    plt.scatter(x, y, c='rgb'[i])
plt.show()
```

```python
session = tf.Session()
```

```python
x1 = df['Sepal.Length'].as_matrix().astype('float32')
x2 = df['Sepal.Width'].as_matrix().astype('float32')
x1, x2
```

```python
features = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']
x = df[features].as_matrix().astype('float32')
x
```

```python
x -= x.min()
x /= x.max()
x
```

```python
session.run(tf.one_hot(labels, 3))
```

```python
w = tf.Variable(tf.truncated_normal([len(features), len(uniques)]))
b = tf.Variable(tf.truncated_normal([len(uniques)]))
y = tf.nn.softmax(tf.matmul(x, w) + b)

z = tf.one_hot(labels, 3)

cross_entropy = -tf.reduce_sum(z * tf.log(y), axis=1)
loss = tf.reduce_mean(cross_entropy)

train = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
```

```python
session.run(tf.cast(tf.equal(1, 1), tf.float32))
```

```python
session.run(tf.cast(tf.equal(1, 2), tf.float32))
```

```python
correct = tf.equal(tf.argmax(y, axis=1), labels)
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```

```python
session.run(tf.global_variables_initializer())

def draw(index):
    print(index, session.run(accuracy))
    c = session.run(tf.argmax(y, axis=1))
    plt.scatter(x[:, 0], x[:, 1], c=['rgb'[color] for color in c])
    plt.show()

draw(0)

for i in range(100):
    for j in range(5):
        session.run(train)
    draw(i + 1)
```
