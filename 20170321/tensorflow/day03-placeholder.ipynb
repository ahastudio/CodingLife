```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt

# pip3 install tensorflow
# pip3 install pandas
# pip3 install matplotlib
```

```python
df = pd.read_csv('./iris.csv')
df
```

```python
labels, uniques = pd.factorize(df['Species'])
labels, uniques
```

```python
session = tf.Session()
```

```python
features = ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']
x_data = df[features].as_matrix().astype('float32')
x_data
```

```python
x_data -= x_data.min()
x_data /= x_data.max()
x_data
```

```python
t_data = session.run(tf.one_hot(labels, 3))
t_data
```

```python
x = tf.placeholder(tf.float32, [None, len(features)])
t = tf.placeholder(tf.float32, [None, len(uniques)])

w = tf.Variable(tf.truncated_normal([len(features), len(uniques)]))
b = tf.Variable(tf.truncated_normal([len(uniques)]))
y = tf.nn.softmax(tf.matmul(x, w) + b)

cross_entropy = -tf.reduce_sum(t * tf.log(y), axis=1)
loss = tf.reduce_mean(cross_entropy)

train = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
```

```python
correct = tf.equal(tf.argmax(y, axis=1), labels)
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```

```python
session.run(tf.global_variables_initializer())

data = {
    x: x_data,
    t: t_data
}

def draw(index):
    print(index + 1, session.run(accuracy, feed_dict=data))
    c = session.run(tf.argmax(y, axis=1), feed_dict=data)
    plt.scatter(x_data[:, 0], x_data[:, 1], c=['rgb'[color] for color in c])
    plt.show()

draw(0)

for i in range(100):
    for j in range(5):
        session.run(train, feed_dict=data)
    draw(i + 1)
```
